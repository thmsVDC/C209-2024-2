```python
import moviepy.editor as mpe
```

pegando o vídeo e salvando em uma variável

```python
video = mpe.VideoFileClip("video_projeto.mp4")
```

Mostrando alguns elementos do vídeo

```python
print('Tamanho:', video.size)
print('FPS:', video.fps)
print('Duração:', video.duration, 'segundos')
print('Número de frames:', video.reader.nframes)
```

# Questão 1

Este código divide o vídeo original em vários clipes de 20 segundos. Para cada clipe cuja posição é par, de 20 em 20 segundos, é aplicado um efeito de espelhamento horizontal. Em seguida, os clipes são concatenados para formar um novo vídeo.

```python
parts = []
for i in range(int(video.duration / 20)):
    subclip = video.subclip(i * 20, i * 20 + 20)
    if i % 2 == 0:
        subclip = subclip.fx(mpe.vfx.mirror_x)

    parts.append(subclip)

questao1 = mpe.concatenate_videoclips(parts)
questao1.write_videofile('invertindo_20_segundos.mp4')
```

# Questão 2

O código processa o vídeo gerado na Questão 1, dividindo-o em partes de 30 segundos cada. Nessas partes, o volume é reduzido para 20%. Por fim, uma seção adicional de 10 segundos tem o volume zerado antes de o vídeo final ser salvo.

```python
parts = []
for i in range(int(questao1.duration / 30)):
    subclip = video.subclip((questao1.duration - 30) - (i * 30), questao1.duration - (i * 30))
    subclip = subclip.volumex(0.2)
    parts.append(subclip)

audio_decrease_video = mpe.concatenate_videoclips(parts)
part = audio_decrease_video.subclip(int(audio_decrease_video.duration - 10), audio_decrease_video.duration)
part = part.volumex(0)
questao2 = mpe.concatenate_videoclips([audio_decrease_video, part])
questao2.write_videofile('audio_diminuindo_30_segundos.mp4')
```

# Questão 3

O código corta o vídeo da Questão 2 em três partes específicas: os primeiros 60 segundos, o intervalo entre 80 segundos e o final do vídeo, e o trecho entre 60 e 80 segundos. Depois, essas partes são reorganizadas e concatenadas para formar um novo vídeo final.

```python
part1 = questao2.subclip(0, 60)
part2 = questao2.subclip(80, questao2.duration)
part3 = questao2.subclip(60, 80)

questao3 = mpe.concatenate_videoclips([part1, part2, part3])
questao3.write_videofile('corte_de_60_a_80.mp4')
```
